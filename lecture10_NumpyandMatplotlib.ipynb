```python
import numpy as np
```

```python
x = np.linspace(-2*np.pi, 2*np.pi, 100)
y = np.linspace(-np.pi, np.pi, 50)
xx, yy = np.meshgrid(x, y)
xx.shape, yy.shape
f = np.sin(xx) * np.cos(0.5*yy)
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
plt.pcolormesh(f)
```
